El área superficial horizontal $A_s$ (m²) de un lago a una profundidad particular se puede calcular a partir del volumen por diferenciación:

$$A_s(z) = -\frac{dV(z)}{dz}$$

donde:
- $V$ = volumen ($m^3$)
- $z$ = profundidad (m) medida desde la superficie hacia el fondo.


La concentración promedio de una sustancia que varía con la profundidad $c$ ($g/m^3$) se puede calcular mediante integración: 

$$c = \frac{\int_0^Z c(z)A_s(z) dz}{V_T}$$

donde:
- $Z$ = profundidad total (m)
- $V_T$ = volumen total del lago, dado también por:


$$V_T = \int_0^Z A_s(z) dz$$

La regla de Boole y la regla de Weddle son métodos de integración numérica que usan aproximaciones polinomiales de orden superior.

**Regla de Boole**
Usa 5 puntos (polinomio de grado 4):
$$
\int_a^b f(x)dx \approx \frac{2h}{45}[7f(x_0) + 32f(x_1) + 12f(x_2) + 32f(x_3) + 7f(x_4)]
$$
donde $h = \frac{b-a}{4}$ y $x_i = a + i\cdot h$


**Regla de Weddle**
Usa 7 puntos (polinomio de grado 6):
$$
\int_a^b f(x)dx \approx \frac{3h}{10}[f(x_0) + 5f(x_1) + f(x_2) + 6f(x_3) + f(x_4) + 5f(x_5) + f(x_6)]
$$
donde $h = \frac{b-a}{6}$ y $x_i = a + i\cdot h$
