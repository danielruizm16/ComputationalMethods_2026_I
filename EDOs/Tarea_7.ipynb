Considera la siguiente ecuación diferencial de segundo orden:
$$
\frac{d^2 x}{dt^2} = f\left(t,x,\frac{dx}{dt}\right),
$$
que puede descomponerse en un sistema acoplado de ecuaciones de primer orden:
$$
\frac{dx}{dt} = v, \quad \frac{dv}{dt} = f(t,x,v).
$$

El método Leapfrog actualiza las variables en pasos de tiempo intercalados ($\Delta t$) usando las siguientes fórmulas:
$$
v_{n+3/2} = v_{n+1/2} + f(t_n, x_n, v_{n+1/2}) \Delta t,
$$
$$
x_{n+1} = x_n + v_{n+1/2} \Delta t.
$$

**Nota:** La siguiente relación puede ser útil
$$
v_{n+1/2}=v_{n}+\frac{1}{2}f(t_n, x_{n}, v_n)\Delta t
$$
