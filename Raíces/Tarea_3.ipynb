Con base en los siguientes pasos, escriba una función que calcule la raíz de una función. Pruébela para encontrar la primera raíz positiva de $f(x)=\sin{(x)} +\cos{(1+x^2)}-1$. Examine la convergencia en función de la tolerancia.

### Pasos


1. **Inicialización:**
    Comience con un intervalo $[a, b]$ donde $f(a)$ y $f(b)$ tengan signos opuestos.
    - Comience con una suposición inicial $x0$ cercana a la raíz sospechada.

2. **Fórmula de Iteración:**
    - En cada iteración, evalúe $ f(x_n) $, $ f'(x_n) $ y $ f''(x_n) $.
    - Calcule la siguiente aproximación $x_{n+1}$ usando la fórmula:
    $$x_{n+1} = x_n - \frac{2f(x_n)f'(x_n)}{2[f'(x_n)]^2 - f(x_n)f''(x_n)}$$
    Esta fórmula se deriva de la expansión de Taylor de segundo orden y proporciona una convergencia cúbica.

3. **Verificación de Convergencia:**
   - Evalúe $ f(x_{n+1}) $.
   - Si $ |f(x_{n+1})| < \text{tolerancia} $ o $ |x_{n+1} - x_n| < \text{tolerancia} $, entonces $ x_{n+1} $ es la raíz aproximada.
   - De lo contrario, establezca $ x_n = x_{n+1} $ y repita el proceso.

5. **Criterios de Parada:**:
   - La iteración se detiene cuando el valor de la función está suficientemente cerca de cero o el cambio en la raíz aproximada es menor que una tolerancia especificada.
